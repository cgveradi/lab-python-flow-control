In the last lab, you were starting an online store that sells various products. To ensure smooth operations, you developed a program that manages customer orders and inventory.

Follow the steps below to complete the exercise:

2. Instead of asking the user to input the name of three products that a customer wants to order, do the following:
    
    a. Prompt the user to enter the name of a product that a customer wants to order.
    
    b. Add the product name to the "customer_orders" set.
    
    c. Ask the user if they want to add another product (yes/no).
    
    d. Continue the loop until the user does not want to add another product.

3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for  the products that were ordered (those in "customer_orders").

```python
# 1. Define a list of products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# 2. Create an empty dictionary for inventory
inventory = {}

# 3. Ask the user for quantity of each product
print("Enter the quantity available for each product:")
for product in products:
    quantity = input(f"Quantity of {product}: ")

    # Ensure user enters a valid number
    while not quantity.isdigit():
        print("Please enter a valid number.")
        quantity = input(f"Quantity of {product}: ")

    inventory[product] = int(quantity)


# 4. Create an empty set for customer orders
customer_orders = set()

# 5. Ask the user to enter product orders repeatedly
print("\nEnter the products that the customer wants to order.")
while True:
    order = input("Enter a product name: ").lower()

    # Validate input
    while order not in products:
        print("Invalid product. Please enter a valid product from the list.")
        order = input("Enter a product name: ").lower()

    customer_orders.add(order)

    # Ask if they want to add another product
    another = input("Add another product? (yes/no): ").lower()
    if another != "yes":
        break


# 6. Print the products ordered
print("\nCustomer has ordered these products:")
for item in customer_orders:
    print("-", item)


# 7. Calculate statistics
total_products_ordered = len(customer_orders)
percentage_ordered = (total_products_ordered / len(products)) * 100

# 8. Store statistics in a tuple
order_status = (total_products_ordered, percentage_ordered)

# 9. Print order statistics
print("\nOrder Statistics:")
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")


# 10. Update inventory for only ordered products
for item in customer_orders:
    inventory[item] -= 1

# 11. Print updated inventory
print("\nUpdated Inventory:")
for product, qty in inventory.items():
    print(f"{product}: {qty}")
```
